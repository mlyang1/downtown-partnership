```python
import pandas as pd
df = pd.read_csv('../data/monthly_totals.csv')
df
```

```python
t = df.set_index('Year').stack().reset_index()
t.columns = ['year','month','count']
t['date'] = pd.to_datetime(t.apply(lambda r: f"1 {r.month} {r.year}", axis=1))
t[['date','count']].to_csv('../data/monthly_totals.csv')
```

```python
import pandas as pd
df = pd.read_csv('../data/neighborhood_totals.csv')
df.columns = [e.lower() for e in df.columns]

df['date'] = pd.to_datetime(df.apply(lambda r: f"1 {r.month} {r.year}", axis=1))
df = df.drop(columns=['year','month'])
cols = [list(df.columns)[-1]] + list(df.columns)[:-1]
df = df[cols]
df.to_csv('../data/neighborhood_totals.csv')
```
